```python
dataset = pd.read_csv('creditcard.csv')
```

```python
df = dataset.sample(n = 10000,random_state = 42)
print(df[df['Class'] == 1]['Class'].shape,df[df['Class'] == 0]['Class'].shape,df.shape)
```

```python
x = df.drop('Class',axis = 1)
y = df['Class']
x
```

```python
print(x.shape,y.shape)
```

0 = Real 1 = Fraud

```python
import numpy as np
df[df['Class'] == 0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
print(x_train.shape,y_train.shape)
```

##Handling Imbalanced Dataset

```python
#clearing datasets finding any nan values
print(np.isnan(y_train).sum())     #returns no of rows with nan  np.isnan(y_train)-> converts y_train to a boolean array
print(np.where(np.isnan(y_train))) #print(y_train.iloc[6326]) = NaN
print(np.where(np.isnan(x_train)))
print(x_train.iloc[6326,:])
```

```python
from sklearn.impute import SimpleImputer        #input = 2D array in numpy or Dataframe
si = SimpleImputer(strategy = 'most_frequent')
si.fit(pd.DataFrame(y_train))   #[[series]] = dataframe
y_train = si.transform(pd.DataFrame(y_train))
y_train = pd.Series(y_train.ravel(),name = 'Class')
```

```python
print(np.isnan(y_train).sum())
print(np.isnan(x_train).sum())
```

```python
print(type(y_train),y_train.shape)
```

```python
print(y_train.iloc[:56])
```

###Oversampling the Minority

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy = 'minority')
X_train,Y_train = smote.fit_resample(x_train,y_train)     #smote cant be applied with NaN values
```

```python
from imblearn.over_sampling import ADASYN
ada = ADASYN(sampling_strategy = 'minority')
X_train_ad,Y_train_ad = ada.fit_resample(x_train,y_train)
```

```python
print(X_train_ad.shape,Y_train_ad.shape)
print(Y_train_ad[Y_train_ad == 1].sum())
```

```python
print(X_train.shape,Y_train.shape)
print(Y_train[Y_train == 1].sum())
```

###Undersampling the Majority

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy = 'majority')
X_train_1,Y_train_1 = rus.fit_resample(x_train,y_train)
```

```python
print(X_train_1.shape,Y_train_1.shape)
```

```python
print(X_train.iloc[:5,:])
print(Y_train[:5])
```

##Using SVM

###Scaling the Predictors

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_svc = sc.fit_transform(X_train)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf',random_state = 42,class_weight='balanced')
classifier.fit(X_train_svc,Y_train)
```

```python
print(np.isnan(x_test).sum())
```

```python
Y_pred_svc = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,Y_pred_svc)
print(cm)
ac = accuracy_score(y_test,Y_pred_svc)
print(ac)
```

```python
print(Y_pred_svc[Y_pred_svc == 1].sum())
print(y_test[y_test == 1].sum())
```

```python
X_train_ad_svc = sc.fit_transform(X_train_ad)
classifier1 = SVC(kernel = 'rbf',random_state = 42,class_weight='balanced')
classifier1.fit(X_train_ad_svc,Y_train_ad)
```

```python
Y_pred_ad_svc = classifier1.predict(x_test)
cm = confusion_matrix(y_test,Y_pred_ad_svc)
print(cm)
ac = accuracy_score(y_test,Y_pred_ad_svc)
print(ac)
```

#Using RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(random_state = 42,class_weight = 'balanced')
classifier_rf.fit(X_train,Y_train)
```

```python
Y_pred_rf = classifier_rf.predict(x_test)
cm = confusion_matrix(y_test,Y_pred_rf)
print(cm)
ac = accuracy_score(y_test,Y_pred_rf)
print(ac)
```
